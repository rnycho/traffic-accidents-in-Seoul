```python
import pandas as pd

years = list(range(2014,2024))
cols = []
accident = []

for i in range(len(years)):
    file = f'./도로교통공단_시도 시군구별 교통사고({years[i]}).csv'
    df = pd.read_csv(file, encoding = 'cp949')
    df_temp = df.head(25)
    accident.append(list(df_temp['사고건수'].values))

cols = df_temp['시군구']

df_final = pd.DataFrame(accident, columns=cols, index=years)

a_df = df_final.T
a_df
```

```python
b_df = a_df[a_df.index == '강남구']
```

```python
df_final.columns
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')

plt.figure(figsize=(10,10))
x = df_final.index
y = df_final.values
plt.plot(x, y, label = df_final.columns)
plt.title('Traffic Accident in Seoul')
plt.xlabel('years')
plt.ylabel('count')
plt.legend(loc='upper right')
plt.show()
```

```python
df = pd.read_csv('./seoulpop.csv', encoding = 'cp949')
d_df = df
e_df = d_df[d_df['연령별'] == '합계']
f_df = e_df[e_df['항목'] == '계']
g_df = f_df[['2014. 4/4', '2015. 4/4','2016. 4/4', '2017. 4/4', '2018. 4/4',
    '2019. 4/4', '2020. 4/4','2021. 4/4', '2022. 4/4', '2023. 4/4']]

g_df[1:]
```

```python
acc_pop_list = b_df.values / g_df[1:].values
acc_pop_list
```

```python
years = range(2014, 2024)
districts = ['종로구', '중구', '용산구', '성동구', '동대문구', '성북구', '도봉구', '은평구', '서대문구', '마포구',
       '강서구', '구로구', '영등포구', '동작구', '관악구', '강남구', '강동구', '송파구', '서초구', '양천구',
       '중랑구', '노원구', '광진구', '강북구', '금천구']

df_final_final = pd.DataFrame(acc_pop_list, columns=years, index=districts)
df_final_final
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
plt.figure(figsize=(10,10))
plt.title('Traffic Accident in Seoul')
plt.xlabel('years')
plt.ylabel('rate')

colors = ["#E6194B", "#3CB44B", "#FFE119", "#4363D8", "#F58231",
          "#911EB4", "#46F0F0", "#F032E6", "#BCF60C", "#FABEBE",
          "#008080", "#E6BEFF", "#9A6324", "#FFFAC8", "#800000",
          "#AAFFC3", "#808000", "#FFD8B1", "#000075", "#808080",
          "#000000", "#A9A9A9", "#FF69B4", "#2E8B57", "#B22222"]

for i, district in enumerate(districts):
    plt.plot(years, df_final_final.loc[district], label=district, color=colors[i])
    plt.legend(loc='upper right')

plt.show()
```
